```python
import requests
from tqdm import tqdm

def download_video(url, save_path):
    # 发送请求，获取响应对象
    response = requests.get(url, stream=True)
    # 获取视频资源大小
    total_size = int(response.headers.get('content-length', 0))
    # 初始化下载大小
    downloaded_size = 0
    # unit: 下载单位（字节）
    unit = 'B'
    unit_scale = 1
    unit_divisor = 1024

    # 保存文件
    with open(save_path, "wb") as f:
        # 创建进度条
        with tqdm(total=total_size, unit='B', unit_scale=unit_scale, unit_divisor=unit_divisor, desc="Downloading") as bar:
            # 循环读取响应内容
            for data in response.iter_content(chunk_size=1024*1024):
                # 写入文件
                f.write(data)
                # 更新下载大小
                downloaded_size += len(data)
                # 更新进度条
                bar.update(len(data))
    print("下载完成...")

# 调用下载函数
video_url = "https://v26-web.douyinvod.com/23e00dfa6332739a65650182ebef61a0/67e23491/video/tos/cn/tos-cn-ve-15/ocIZA8aCiPIiBABEA5TFgPALoIiUQAAKdYxjQ/?a=6383&ch=10010&cr=3&dr=0&lr=all&cd=0%7C0%7C0%7C3&cv=1&br=663&bt=663&cs=2&ds=3&ft=AJkeU_TLRR0sTHC4NDv2Nc.xBiGNbLLaZq1U_4q6SOKSNv7TGW&mime_type=video_mp4&qs=15&rc=NGU4OWk1OmdmOjczNmk2O0BpajVpNXM5cjM5eTMzNGkzM0BjNTYzMzMuNl4xX2JhMGA0YSMvMGZeMmQ0ZC1gLS1kLS9zcw%3D%3D&btag=80000e00018000&cquery=100o_101r_100B_100x_100z&dy_q=1742867010&feature_id=d25ea033481b28b23c6860d19885cf8f&l=202503250943297F01C20DEB642534AC9E&__vid=7476089740383538444"
path = r"F:\Jupyter\crawler\video.mp4"
download_video(video_url, path)
```

```python
import requests
from tqdm import tqdm

def download_video(url, save_path):
    # 添加请求头模拟浏览器
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
        'Referer': 'https://www.douyin.com/',
        'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
        'Accept-Language': 'zh-CN,zh;q=0.8,zh-TW;q=0.7,zh-HK;q=0.5,en-US;q=0.3,en;q=0.2',
    }
    
    # 发送请求，获取响应对象
    response = requests.get(url, headers=headers, stream=True)
    
    # 检查响应状态码
    if response.status_code != 200:
        print(f"下载失败，状态码：{response.status_code}")
        print(f"响应头：{response.headers}")
        return
    
    # 获取视频资源大小
    total_size = int(response.headers.get('content-length', 0))
    
    # 保存文件
    with open(save_path, "wb") as f:
        # 创建进度条
        with tqdm(total=total_size, unit='B', unit_scale=True, unit_divisor=1024, desc="Downloading") as bar:
            # 循环读取响应内容
            for data in response.iter_content(chunk_size=1024*1024):
                # 写入文件
                f.write(data)
                # 更新进度条
                bar.update(len(data))
    
    print("下载完成...")

# 调用下载函数
video_url = "https://v3-web.douyinvod.com/b69b63d9bfad3c6ee683096ade0942c4/67e391fa/video/tos/cn/tos-cn-ve-0015c800/oUw41fD4QFAXIHBOqgECZoAFApcVf9gznp4xgA/?a=6383&br=895&bt=895&btag=c0000e00010000&cd=0%7C0%7C0%7C3&ch=26&cquery=100x_100z_100o_100w_100B&cr=3&cs=0&cv=1&dr=0&ds=6&dy_q=1742956466&feature_id=46a7bb47b4fd1280f3d3825bf2b29388&ft=AJkeU_TLRR0sTHC4NDv2Nc.xBiGNbLQH~q1U_44MSOKSNv7TGW&is_ssr=1&l=202503261034264ABF452D76BFC6035AB4&lr=all&mime_type=video_mp4&qs=12&rc=NjY7aTs8OThpNDhpOTVoNUBpajM0OG45cnh2djMzNGkzM0AzLTU2YzI1NmExNTJgXjNeYSNebDNwMmQ0X3BgLS1kLS9zcw%3D%3D&__vid=7438194471507856651"
path = r"F:\Jupyter\crawler\yong.mp4"
download_video(video_url, path)
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.dpm.org.cn/lights/royal.html"
try:
    req = requests.get(url)
    req.raise_for_status()
    soup = BeautifulSoup(req.text, "html.parser")
    pic1 = soup.find("img")
    if pic1 is not None:
        # 这里原代码用的title属性，但示例HTML里是alt属性，使用alt更合适
        imgmz = pic1.get('alt')
        imglj = pic1.get('src')
        print(imglj)
        print(imgmz)
        with open("tpxx.txt", "w", encoding="utf-8") as f1:
            if imglj:
                f1.write(imglj + "\n")
            if imgmz:
                f1.write(imgmz + "\n")
    else:
        print("未在页面中找到img标签。")
except requests.RequestException as e:
    print(f"请求出错: {e}")
except Exception as e:
    print(f"发生未知错误: {e}")
```

```python
import requests
headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
    }

url='https://webapi.qingting.fm/api/mobile/rank/hotSaleWeekly'

response = requests.get(url,headers=headers).json()#拿到网页转换成Json形式
# print(response)
capacity=response.get("rankinglist")#获取字典
# print(capacity)
result_dict={}
j=1
for i in capacity:
    #将数据写入字典
    title=i.get("title")
    desc=i.get("desc")
    playCount=i.get("playCount")
    imgUrl=i.get("imgUrl")
    one_dict={}
    one_dict["title"]=title
    one_dict["playCount"]=playCount
    one_dict["desc"]=desc
    one_dict["imgUrl"]=imgUrl
    result_dict[j]=one_dict  
    j+=1
# print(result_dict)
#打印字典
for  key, value in result_dict.items():
    print(f"book:{key}   content: {value}")


```

```python
from bs4 import BeautifulSoup

html = """
<!DOCTYPE html>
<html>
<head>
    <title>示例页面</title>
</head>
<body>
    <div id="main" class="content">
        <h1>这是标题</h1>
        <p>这是段落内容。</p>
        <a href="https://www.example.com" target="_blank">示例链接</a>
    </div>
</body>
</html>
"""

# 创建 BeautifulSoup 对象
soup = BeautifulSoup(html, 'html.parser')

# tag.name 属性：获取标签名称
div_tag = soup.div
print("div 标签的名称:", div_tag.name)

# tag.string 属性：获取标签内的文本内容
h1_tag = soup.h1
print("h1 标签的文本内容:", h1_tag.string)

# tag.attrs 属性：获取标签的所有属性
a_tag = soup.a
print("a 标签的所有属性:", a_tag.attrs)

# 单独获取某个属性
print("a 标签的 href 属性:", a_tag['href'])
```
